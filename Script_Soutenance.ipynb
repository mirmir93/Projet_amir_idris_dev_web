## SLIDE 2 — Introduction au Projet ⏱ 1min30

🟧 **IDRIS :**
> Notre idée c'était de faire un site où on peut chercher n'importe quel pays du monde, le filtrer par continent, et voir toutes ses informations détaillées comme la population, les langues, les monnaies, le drapeau, etc.

> Techniquement, on a voulu faire ce qu'on appelle une Single Page Application. Ça veut dire que tout se passe sur une seule page HTML. Quand on clique sur un pays par exemple, la page ne se recharge pas — c'est React qui met à jour juste la partie qui change. C'est beaucoup plus fluide pour l'utilisateur.

> Pour faire ça, on a utilisé les hooks modernes de React : `useState` pour stocker les données qui changent, et `useEffect` pour charger les pays depuis l'API au démarrage.

## SLIDE 3 — Fonctionnalités Principales ⏱ 2min

🟦 **AMIR :**
> On a 5 fonctionnalités principales dans notre application.

> La première c'est la recherche en temps réel. Quand on tape une lettre dans la barre de recherche, les résultats se filtrent instantanément. Et ça marche que ce soit en français, en anglais ou même dans la langue du pays. Par exemple si on tape "Deutschland" ça va trouver l'Allemagne. En plus on a géré le problème des accents : si on tape "Etats" sans accent, ça trouve quand même "États-Unis". Pour ça on a codé une fonction qui nettoie le texte avant de comparer.

🟧 **IDRIS :**
> La deuxième fonctionnalité c'est le filtre par continent. On a un menu déroulant avec Afrique, Amériques, Asie, Europe, Océanie. Et ce qui est bien c'est que les deux filtres se combinent. Par exemple on peut taper "France" et en même temps choisir "Europe" — le filtre fait un ET logique.

> La troisième c'est la vue détaillée. Quand on clique sur un pays, on voit tout : nom natif, population, sous-région, langues, monnaies, frontières, et le drapeau en grand. On a aussi ajouté un lien Wikipedia pour que l'utilisateur puisse en savoir plus. C'est la quatrième fonctionnalité.

> Et la cinquième c'est que le site est responsive. Il s'adapte tout seul au mobile, à la tablette et au PC. On n'a pas eu besoin de coder chaque taille d'écran séparément.

## SLIDE 4 — Technologies Employées ⏱ 2min

🟦 **AMIR :**
> Comme framework on a choisi React. Ce qui est bien avec React c'est qu'il utilise un DOM virtuel. En gros au lieu de redessiner toute la page quand quelque chose change, il compare l'ancien et le nouveau rendu et il met à jour seulement ce qui a changé. C'est ça qui rend l'appli rapide.

> Et pour l'outil de build on a pris Vite au lieu de l'ancien Create React App. Vite est beaucoup plus rapide : le serveur de dev démarre en moins d'une seconde, et quand on modifie un fichier le changement apparaît tout de suite dans le navigateur grâce au Hot Module Replacement.

🟧 **IDRIS :**
> Pour les données des pays, on utilise l'API REST Countries. C'est une API gratuite et publique qui nous donne toutes les infos sur tous les pays du monde. Mais on a optimisé nos requêtes : au lieu de tout télécharger, on a mis `?fields=` dans l'URL pour demander uniquement ce dont on avait besoin — le nom, le drapeau, la population, la région, les langues, etc. Ça réduit le temps de chargement.

> Pour le CSS on a utilisé CSS Grid et Flexbox. La grille utilise `auto-fill` avec `minmax(260px, 1fr)` — concrètement ça fait que les cartes se réorganisent automatiquement : 4 par ligne sur PC, 2 sur tablette, 1 sur mobile. Et pour les icônes dans l'application on a pris Lucide-React, une bibliothèque légère.

## SLIDE 5 — Architecture du Code ⏱ 1min30

🟦 **AMIR :**
> Toute l'appli repose sur un seul composant principal qui s'appelle App. Ce composant gère 4 variables d'état avec useState :
> 
> - `countries` c'est la liste de tous les pays qu'on reçoit de l'API
> - `searchTerm` c'est ce que l'utilisateur tape dans la barre de recherche
> - `selectedRegion` c'est le continent choisi dans le menu déroulant
> - `selectedCountry` c'est le pays sur lequel on a cliqué pour voir les détails
> 
> Dès qu'une de ces variables change, React redessine automatiquement l'interface.

🟧 **IDRIS :**
> Le démarrage de l'appli c'est simple : on a un fichier `index.html` avec juste une div vide dedans, qui a l'id `root`. Ensuite le fichier `index.jsx` utilise `ReactDOM.createRoot` pour prendre le contrôle de cette div et injecter notre composant App dedans. Et après c'est App qui gère tout : il charge les données, filtre les pays, et bascule entre la vue en grille et la vue détail avec un rendu conditionnel — en gros un `if/else` mais en version React, avec l'opérateur ternaire.

## SLIDE 6 — Le Cycle de Vie de l'Application ⏱ 2min

🟧 **IDRIS :**
> L'appli fonctionne en 4 étapes.

> Étape 1 c'est le démarrage. React prend le contrôle de la page HTML avec createRoot.

> Étape 2 c'est l'appel API. On utilise le hook useEffect avec un tableau vide en deuxième argument. Ce tableau vide c'est très important : il dit à React "lance ce code une seule fois au tout début, et c'est tout". Si on met pas ce tableau vide, à chaque fois qu'on tape une lettre dans la recherche, React relance l'appel API et ça fait une boucle infinie. Donc le fetch vers REST Countries se fait une seule fois.

🟦 **AMIR :**
> Étape 3 c'est le filtrage. On a une fonction normalizeText qui nettoie le texte : elle enlève les accents avec normalize NFD, elle supprime les caractères spéciaux avec une expression régulière, et elle met tout en minuscules. Comme ça quand on compare "ÉTATS" avec "etats" ça matche. Ensuite on combine le filtre texte et le filtre continent avec un ET logique : le pays doit correspondre aux deux pour s'afficher.

> Étape 4 c'est l'affichage. On utilise un opérateur ternaire : si selectedCountry est pas null, ça veut dire qu'on a cliqué sur un pays, donc on affiche ses détails. Sinon on affiche la grille avec toutes les cartes. Tout ça sans recharger la page, c'est le principe de la Single Page Application.

## SLIDE 7 — Problèmes Rencontrés & Solutions ⏱ 2min

🟦 **AMIR :**
> On a eu 4 problèmes principaux pendant le développement.

> Le premier c'était la performance. L'API nous renvoyait énormément de données pour chaque pays, même des trucs qu'on utilisait pas. Ça ralentissait le chargement. La solution ça a été d'ajouter `?fields=` dans l'URL pour demander seulement les champs dont on avait besoin : le nom, le drapeau, la population, la région, etc.

> Le deuxième problème c'était les accents. Quand on tapait "Etats" sans accent, ça trouvait pas "États-Unis". Donc on a créé la fonction normalizeText qui décompose les accents avec `normalize("NFD")` et les supprime avec une regex. Après ça, "Etats" et "États" c'est pareil pour la recherche.

🟧 **IDRIS :**
> Le troisième problème c'est que certains pays ont pas de capitale. Par exemple l'Antarctique ou Macao. Quand on essayait d'afficher la capitale, l'appli plantait parce que la donnée existait pas. On a résolu ça avec le chaînage optionnel — c'est le `?.` en JavaScript — qui vérifie si la donnée existe avant d'y accéder. Et si elle existe pas, on affiche "N/A" à la place.

> Le quatrième problème c'est que l'API renvoie tout en anglais. Nous on voulait un site en français. Donc on a fait des dictionnaires de traduction manuels : un objet JavaScript qui dit "quand l'API dit Africa, moi j'affiche Afrique", "quand elle dit French, moi j'affiche Français", etc. C'est un peu artisanal mais ça marche bien.

## SLIDE 8 — Pistes d'Amélioration ⏱ 1min30

🟧 **IDRIS :**
> Si on avait eu plus de temps, on aurait amélioré plusieurs choses.

> D'abord un spinner de chargement. Parce que quand l'appli démarre, il y a un petit moment où la page est vide le temps que l'API réponde. Un spinner ou une animation de chargement ça rassurerait l'utilisateur.

> Ensuite une vraie gestion d'erreurs. Si l'API est en panne, pour l'instant l'utilisateur voit juste une page vide. Il faudrait afficher un message genre "Service indisponible, réessayez plus tard".

🟦 **AMIR :**
> On aurait aussi aimé mettre React Router. Ça permettrait d'avoir des URLs propres comme `/pays/france` au lieu que tout soit sur la même URL. L'avantage c'est qu'on pourrait partager un lien direct vers un pays.

> Un mode sombre aussi ça serait bien, avec un bouton pour switcher. Et enfin un système de favoris sauvegardé dans le localStorage du navigateur, comme ça l'utilisateur retrouve ses pays préférés même après avoir fermé la page.

## SLIDE 9 — Résultats et Compétences ⏱ 1min

🟦 **AMIR :**
> En résultat final, l'appli couvre plus de 250 pays, elle a 5 fonctionnalités complètes, et elle est 100% responsive sur tous les écrans.

🟧 **IDRIS :**
> Ce projet nous a permis de maîtriser les bases de React : les hooks useState et useEffect, comment consommer une API REST avec fetch, et le rendu conditionnel. Et aussi des compétences en CSS avec Grid et Flexbox pour le design responsive, la gestion d'état, et le debugging quand quelque chose marchait pas.
