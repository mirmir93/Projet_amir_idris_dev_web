# Plan de Soutenance : Explorateur de Pays

> Ce plan est prévu pour une présentation d'environ **10 à 15 minutes** + questions du jury. Adapte la durée selon ce qu'on t'a demandé.

## PARTIE 1 — Introduction (~2 min)

**Ce que tu dis :**
> *"Bonjour, je m'appelle [Prénom Nom], et aujourd'hui je vais vous présenter mon projet de développement web : un **Explorateur de Pays** interactif, développé en React."*

- **Le contexte** : Expliquer en une phrase le but du projet → *"L'objectif était de créer une application web permettant de rechercher, filtrer et consulter les informations détaillées de n'importe quel pays du monde."*
- **Les fonctionnalités principales** (liste rapide) :
  1. Recherche en temps réel (par nom, en français, anglais ou natif)
  2. Filtrage par continent
  3. Vue détaillée ("Biographie") d'un pays avec ses langues, monnaies, frontières
  4. Lien direct vers la page Wikipédia du pays
  5. Interface 100% responsive (mobile + PC)

## PARTIE 2 — Démonstration en direct (~3 min)

> **Ouvre ton application dans le navigateur** (`npm run dev`) et montre concrètement :

1. **La page d'accueil** : Montre la grille de tous les pays, le compteur dynamique ("250 pays trouvés").
2. **La barre de recherche** : Tape "fran" → Montre que la France apparaît en temps réel. Ensuite tape "senegal" (sans accent) → Montre que le Sénégal apparaît quand même grâce à ton `normalizeText`.
3. **Le filtre par continent** : Sélectionne "Asie" → Montre que la grille se réduit instantanément.
4. **Combinaison recherche + filtre** : Tape "in" avec le filtre "Asie" → Seuls les pays d'Asie contenant "in" s'affichent (Inde, Indonésie, Chine...).
5. **La vue Biographie** : Clique sur un pays → Montre le drapeau, la population, les langues traduites en français, les monnaies, les pays frontaliers, et le lien Wikipedia.
6. **Le bouton Retour** : Reviens à la liste pour montrer la navigation sans rechargement.
7. **Le responsive** : Réduis la fenêtre du navigateur → Montre que les cartes passent de 4 colonnes à 1 colonne automatiquement.

## PARTIE 3 — Les Choix Techniques (~3 min)

> **Pourquoi ces technologies ?**

| Technologie | Justification à donner |
|---|---|
| **React** | "React gère le DOM virtuel, ce qui rend l'interface ultra-réactive. Chaque frappe clavier met à jour la liste sans recharger la page." |
| **Vite** | "J'ai choisi Vite plutôt que Create React App car le serveur de développement démarre en moins d'une seconde et le rechargement à chaud (HMR) est instantané." |
| **Lucide-React** | "Bibliothèque d'icônes légère et moderne, par rapport à FontAwesome qui pèse beaucoup plus lourd pour le même résultat." |
| **REST Countries API** | "API publique et gratuite qui fournit les données de tous les pays. J'ai optimisé l'appel avec le paramètre `?fields=` pour ne récupérer que les champs nécessaires." |
| **CSS Grid + Flexbox** | "La grille `auto-fill + minmax` me permet d'avoir un design responsive automatique sans écrire de media queries complexes." |

## PARTIE 4 — L'Architecture du Code (~3 min)

> **Montre le code dans VS Code** (ou ton notebook) et explique la structure :

### Étape 1 : Le démarrage
- *"Le navigateur charge `index.html`, qui contient un `<div id="root">` vide."*
- *"Le fichier `index.jsx` s'active : il utilise `ReactDOM.createRoot` pour injecter le composant `<App />` dans cette div."*

### Étape 2 : Les 4 états (`useState`)
- *"Mon application repose sur 4 variables d'état :"*
  - `countries` → La liste brute de l'API
  - `searchTerm` → Le texte tapé par l'utilisateur
  - `selectedRegion` → Le continent choisi
  - `selectedCountry` → Le pays cliqué (ou `null`)

### Étape 3 : L'appel API (`useEffect`)
- *"Au premier chargement, `useEffect` avec un tableau vide `[]` déclenche un `fetch` vers l'API. Ça ne se fait qu'une seule fois pour ne pas surcharger le serveur."*

### Étape 4 : Le filtrage intelligent
- *"La fonction `normalizeText` retire les accents et met tout en minuscules pour que la recherche soit tolérante. Le filtre combine texte ET continent avec un ET logique (`&&`)."*

### Étape 5 : L'affichage conditionnel
- *"Un opérateur ternaire (`selectedCountry ? ... : ...`) permet de basculer entre la vue grille et la vue biographie sans changer de page HTML. C'est le principe de la Single Page Application."*

## PARTIE 5 — Les Difficultés Rencontrées & Solutions (~2 min)

> Le jury adore cette partie. Ça montre ta maturité de développeur.

| Problème rencontré | Solution apportée |
|---|---|
| L'API renvoyait trop de données → Lenteur | Ajout du `?fields=...` dans l'URL pour ne demander que les champs utiles |
| Les accents cassaient la recherche ("Etats" ≠ "États") | Création de `normalizeText()` avec `normalize("NFD")` + RegEx |
| Certains pays n'ont pas de capitale (Antarctique, Macao) | Mise en place de gardes-fous (`?.`, `\|\| 'N/A'`) pour éviter les crashs |
| L'API renvoie tout en anglais | Création de dictionnaires de traduction manuels (langues, monnaies, régions, sous-régions) |
| Navigation entre la liste et les détails d'un pays | Utilisation du rendu conditionnel React avec `selectedCountry` au lieu d'un routeur complexe |

## PARTIE 6 — Pistes d'Amélioration (~1 min)

> Montrer que tu as du recul sur ton propre travail :

- *"Si j'avais plus de temps, j'aurais ajouté :"*
  1. **Un état de chargement** (`loading`) avec un spinner pendant que l'API répond
  2. **Un message d'erreur visuel** si l'API est en panne (au lieu d'un simple `console.error`)
  3. **React Router** pour avoir de vraies URLs distinctes (`/pays/france`) et pouvoir partager un lien direct
  4. **Un mode sombre** (Dark Mode) avec un bouton toggle
  5. **La sauvegarde de favoris** en `localStorage` pour retrouver ses pays préférés

Je suis maintenant à l'aise pour développer des interfaces web dynamiques et interactives. Merci pour votre attention, je suis prêt pour vos questions."*

## Aide-Mémoire Express (Questions Fréquentes du Jury)

| Question probable | Ta réponse en 1 phrase |
|---|---|
| "C'est quoi un hook ?" | "C'est une fonction spéciale de React qui permet d'ajouter des fonctionnalités (état, effets) à un composant fonction, sans avoir besoin de classes." |
| "C'est quoi le Virtual DOM ?" | "C'est une copie légère du vrai DOM que React garde en mémoire. Quand un état change, React compare l'ancien et le nouveau Virtual DOM, et ne modifie que ce qui a réellement changé sur la page." |
| "Pourquoi `key={index}` dans le `.map()` ?" | "React a besoin d'un identifiant unique par élément de liste pour savoir lequel ajouter, supprimer ou déplacer efficacement lors des re-renders." |
| "C'est quoi une SPA ?" | "Single Page Application : une seule page HTML est chargée, et JavaScript modifie dynamiquement le contenu sans jamais recharger la page." |
| "Différence entre `dependencies` et `devDependencies` ?" | "Les dependencies sont embarquées dans le site final (React, Lucide). Les devDependencies ne servent qu'au développeur (Vite, le plugin React) et ne sont pas envoyées aux visiteurs." |
| "Pourquoi `fetch` et pas `axios` ?" | "Pour une seule requête GET publique sans authentification, `fetch` natif suffit. Axios serait utile pour des requêtes complexes avec tokens ou intercepteurs." |
| "C'est quoi JSX ?" | "C'est une extension de syntaxe qui permet d'écrire du HTML directement dans du JavaScript. Vite le traduit ensuite en `React.createElement()` que le navigateur peut comprendre." |
