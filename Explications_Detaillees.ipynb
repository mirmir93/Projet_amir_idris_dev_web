## Glossaire : Tous les "Petits Mots" Expliqués Simplement

Avant de plonger dans le code, voici **tous les termes techniques** que tu vas croiser, expliqués comme si on discutait entre potes. Si le jury utilise un de ces mots, tu sauras quoi répondre.

| Mot technique | Explication simple |
|---|---|
| **React** | Une boîte à outils JavaScript créée par Facebook. Au lieu de recharger toute la page quand quelque chose change, React ne redessine **que le bout qui a changé**. C'est pour ça que quand tu tapes dans la barre de recherche, c'est instantané. |
| **Composant** | Un bout de page réutilisable. Dans notre projet, `App` est un composant. C'est comme un Lego : tu le construis une fois, et tu peux l'afficher où tu veux. |
| **JSX** | Ça ressemble à du HTML (`<div>`, `<h1>`) mais c'est écrit dans un fichier JavaScript. Le navigateur ne comprend pas le JSX directement, donc Vite le traduit en vrai JavaScript avant de l'envoyer au navigateur. |
| **Hook** | Un outil spécial de React. C'est une fonction qui commence toujours par `use` (comme `useState`, `useEffect`). Ça te donne des super-pouvoirs dans ton composant (mémoire, actions automatiques, etc.). |
| **State (État)** | La "mémoire" de ton composant. C'est une variable spéciale : quand elle change, React **redessine automatiquement** la page pour montrer la nouvelle valeur. Exemple : quand tu tapes une lettre, le state `searchTerm` change, et la liste se met à jour toute seule. |
| **`useState`** | Le hook qui crée un state. `const [truc, setTruc] = useState('valeur de départ')` → `truc` c'est la valeur actuelle, `setTruc` c'est la fonction pour la modifier. |
| **`useEffect`** | Le hook qui dit "fais cette action à un moment précis". Avec `[]` vide à la fin = "fais-le UNE SEULE FOIS quand la page s'ouvre". Sans ça, le code se relancerait à chaque lettre tapée (boucle infinie). |
| **Re-render** | Quand un state change, React **recalcule** ce qui doit s'afficher. C'est pas un "rechargement de page" (la page reste ouverte), c'est juste React qui met à jour le bout d'écran qui a changé. |
| **DOM** | Le **D**ocument **O**bject **M**odel. C'est la structure de ta page HTML vue par le navigateur. Chaque `<div>`, `<h1>`, `<img>` est un "nœud" du DOM. |
| **DOM Virtuel** | Une copie légère du DOM que React garde en mémoire. Quand quelque chose change, React compare l'ancien et le nouveau DOM virtuel, et ne touche que les éléments qui ont réellement changé (au lieu de tout redessiner). |
| **SPA** | **S**ingle **P**age **A**pplication. Une seule page HTML est chargée. Quand tu "changes de page" (ex: tu cliques sur un pays), en réalité la page ne se recharge pas : c'est juste React qui cache un bout et en montre un autre. |
| **API** | **A**pplication **P**rogramming **I**nterface. C'est comme un serveur au restaurant : tu passes ta commande (une URL), et il te rapporte les données (du JSON). Dans notre projet, l'API c'est `restcountries.com` qui nous donne les infos des pays. |
| **REST** | Un style de communication avec une API. En gros : tu envoies une URL avec un "verbe" (GET = "donne-moi des données", POST = "enregistre ça") et l'API te répond. |
| **JSON** | Le format des données que l'API renvoie. Ça ressemble à un objet JavaScript avec des `{ }` et des `[ ]`. Exemple : `{"name": "France", "population": 67000000}` |
| **`fetch()`** | La fonction JavaScript native (déjà dans le navigateur, pas besoin d'installer quoi que ce soit) qui envoie une requête vers une URL (une API) et attend la réponse. |
| **Promesse (Promise)** | Quand tu appelles `fetch()`, la réponse n'arrive pas tout de suite (il faut le temps qu'internet aille chercher les données). Donc `fetch` te donne une "promesse" = "je te donnerai la réponse quand elle sera prête". |
| **`.then()`** | Signifie "ET ENSUITE". `.then(res => res.json())` = "Quand la réponse arrive, ALORS transforme-la en JSON". C'est comme une file d'attente : chaque `.then()` attend que le précédent ait fini. |
| **`.catch()`** | Le filet de sécurité. Si la requête plante (pas internet, API en panne), `.catch()` attrape l'erreur au lieu de faire crasher toute l'appli. |
| **Ternaire** | Un mini `if/else` sur une seule ligne : `condition ? siVrai : siFaux`. Dans notre code : `selectedCountry ? (affiche détails) : (affiche grille)`. |
| **`.map()`** | Transforme un tableau en un autre tableau. Dans React, on l'utilise pour transformer une liste de données (les pays) en une liste d'éléments HTML (les cartes). C'est comme une machine : tu mets 250 pays en entrée, elle te sort 250 `<article>` en sortie. |
| **`.filter()`** | Garde uniquement les éléments d'un tableau qui passent un test. Ex : "Garde uniquement les pays dont le nom contient ce que j'ai tapé ET qui sont en Asie". |
| **`key`** | Quand tu fais un `.map()` en React, chaque élément doit avoir un identifiant unique (`key={index}`). C'est pour que React sache quel élément a changé, a été ajouté ou supprimé sans tout recalculer. |
| **`import`** | "Va chercher ce fichier/cette librairie et amène-le ici". C'est comme un `#include` en C ou un `from X import Y` en Python. |
| **`export default`** | "Rends ce composant/cette fonction disponible pour que d'autres fichiers puissent l'importer". Sans ça, `index.jsx` ne pourrait pas utiliser `App`. |
| **Props** | Des "paramètres" qu'on passe à un composant. Comme les arguments d'une fonction. (On ne les utilise pas beaucoup dans notre projet car on n'a qu'un seul composant.) |
| **NPM** | Le "magasin d'outils" de JavaScript. `npm install` télécharge toutes les librairies listées dans `package.json`. `npm run dev` lance le serveur. |
| **Dépendance** | Un outil externe que ton projet utilise. Ex : React, Lucide. Elles sont listées dans `package.json`. |
| **devDependency** | Un outil qui sert UNIQUEMENT au développeur (toi). Ex : Vite. Quand le site est publié en ligne, ces outils ne sont pas envoyés aux visiteurs. |
| **Bundler** | Un programme qui prend tous tes fichiers (JSX, CSS, JS) et les "emballe" en un seul paquet optimisé pour le navigateur. Vite est ton bundler. |
| **HMR** | **H**ot **M**odule **R**eplacement. Quand tu modifies ton code, la page se met à jour automatiquement sans recharger. Tu vois le changement en temps réel. |
| **Responsive** | Un site qui s'adapte à toutes les tailles d'écran (téléphone, tablette, PC). Dans ton projet, c'est surtout grâce à la ligne CSS `grid-template-columns: repeat(auto-fill, minmax(260px, 1fr))`. |
| **Flexbox** | Un système CSS pour aligner des éléments sur une seule ligne (ou colonne). Utilisé pour mettre la barre de recherche et le filtre côte à côte. |
| **CSS Grid** | Un système CSS pour créer des grilles (lignes ET colonnes). Utilisé pour afficher les cartes des pays. |
| **Media Query** | Une condition CSS qui dit "applique ce style SEULEMENT si l'écran fait plus/moins de X pixels". Ex : `@media (min-width: 900px)` = "seulement sur grand écran". |
| **`hover`** | L'état quand la souris passe au-dessus d'un élément. `.country-card:hover` = "quand la souris survole une carte". |
| **`transition`** | Rend un changement CSS progressif au lieu d'instantané. `transition: transform 0.3s` = "quand le transform change, fais-le en 0.3 secondes au lieu d'un coup sec". |
| **`transform`** | Déplace, tourne ou redimensionne un élément. `translateY(-10px)` = monte de 10 pixels. `translateX(-5px)` = décale à gauche de 5 pixels. |
| **Regex (RegEx)** | **Reg**ular **Ex**pression. Un "patron" de recherche très puissant. `/[\u0300-\u036f]/g` est une regex qui cible tous les accents Unicode pour les supprimer. |
| **Optional Chaining (`?.`)** | `country?.capital` signifie "si `country` existe, accède à `capital`. Sinon, renvoie `undefined` au lieu de faire crasher l'appli". C'est un garde-fou. |
| **`\|\| 'N/A'`** | L'opérateur "OU". `country.capital \|\| 'N/A'` signifie "prends la capitale, mais si elle n'existe pas, affiche 'N/A' à la place". |
| **`console.error()`** | Affiche un message d'erreur dans la Console du navigateur (F12 → Onglet Console). L'utilisateur normal ne le voit pas, c'est pour le développeur. |
| **Nettoyer le texte** | Supprimer les accents, mettre en minuscules, enlever les espaces en trop. Pour que "  ÉTATS  " et "etats" soient considérés comme identiques lors de la recherche. |

## Préparation à la Soutenance : Réponses aux Questions Clés

Cette section est conçue spécialement pour t'aider à justifier tes choix techniques et à répondre sereinement aux questions du jury.

### 1. L'Architecture Globale et l'Écosystème du Projet

**Q: "Sur quoi est basé ton projet ? Qu'as-tu utilisé pour le configurer ?"**
*   **Vite (`vite.config.js`)** : Ton projet utilise **Vite** plutôt que l'ancien "Create React App" (CRA). Vite est un outil de construction (*bundler*) moderne beaucoup plus rapide car il exploite les modules ES natifs du navigateur, offrant un rafraîchissement (HMR) quasi instantané pendant le développement.
*   **React (`package.json`)** : La librairie centrale pour l'interface utilisateur. Le projet utilise **React 19** (la version la plus récente).
*   **Lucide-React** : Bibliothèque choisie pour les icônes (loupe, carte, flèche de retour, etc.) du fait de sa très grande légèreté par rapport à d'autres alternatives volumineuses comme FontAwesome.

### 2. Le Point d'Entrée (`index.html` et `src/index.jsx`)

**Q: "Comment React s'affiche-t-il sur la page HTML ? Comment ça démarre ?"**
*   **`index.html`** : Ce fichier représente la base de notre Single Page Application (SPA). On y trouve la balise `<div id="root"></div>`, qui agit comme réceptacle vide pour l'application.
*   **`src/index.jsx`** : C'est le point de montage. `ReactDOM.createRoot(document.getElementById('root'))` prend le contrôle du `div#root` pour y injecter toute l'interface (notamment le composant principal `<App />`).
*   **`<React.StrictMode>`** : Un outil de débogage de React. En développement, il monte et démonte intentionnellement chaque composant une fois au démarrage pour traquer des effets secondaires indésirables ou des pratiques obsolètes (sans impact en production).

### 3. Le Cœur du Projet : `src/App.jsx`

#### A. La gestion d'états (Hooks `useState`)
**Q: "À quoi servent ces `useState` ? Pourquoi ne pas utiliser des variables classiques ?"**
*   En React, modifier une variable ordinaire (ex: `let mot = ''`) ne met pas à jour l'interface. En utilisant `useState`, tout changement de valeur indique à React qu'il doit **redessiner** l'affichage.

#### B. La Récupération des Données (Hook `useEffect`)
**Q: "Pourquoi faire l'appel API dans un `useEffect` vide ?"**
*   Sans `useEffect`, chaque fois qu'on taperait une lettre dans la recherche, React recommencerait l'appel à l'API (boucle infinie). Le `[]` garantit que l'appel API ne se fait **qu'une seule fois au chargement initial**.

#### C. Le Mécanisme de Filtrage (`filteredCountries`)
**Q: "Comment as-tu codé la barre de recherche ?"**
*   Le state d'origine `countries` n'est jamais modifié. Une nouvelle liste `filteredCountries` est **dérivée**. La condition `matchesSearch && matchesRegion` garantit qu'un pays doit répondre aux deux critères (texte ET continent) pour s'afficher.

#### D. Le Rendu Conditionnel
**Q: "Comment naviguer entre la liste et le détail sans changer de page ?"**
*   Un **ternaire** : `{selectedCountry ? (Vue Détail) : (Vue Grille)}`. Si un pays est cliqué, on montre ses détails. Sinon, on montre la grille.

### 💡 Questions Pièges Possibles

    → "Pour une seule requête GET publique, `fetch` (natif) suffit. Axios serait utile pour des requêtes complexes avec tokens."
